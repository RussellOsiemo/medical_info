```python
import scispacy 
import spacy

#core models
import en_core_sci_sm
import en_core_sci_md

#NER specific models
import en_ner_bc5cdr_md

#extracting and displaying data
import pandas as pd 
from spacy import displacy
```

```python
# get the data
df = pd.read_csv('mtsamples.csv',error_bad_lines=False)
df.head()
```

```python
# pick a sample to test the model and test the scispacy NER model
sample = df.loc[40, "transcription"]

# load a model and pass the text through
nlp_sm = en_core_sci_sm.load()
doc = nlp_sm(sample)

# display the results
# entity extraction
displacy_image = displacy.render(doc, jupyter = True, style= 'ent' )

```

```python
 # Load the specific model: en_core_sci_md and pass text through

nlp_md = en_core_sci_md.load()
doc = nlp_md(sample)
# results of the extraction
displacy_image = displacy.render(doc , jupyter = True, style = "ent")
```

```python
# Now Load specific model: import en_ner_bc5cdr_md and pass text through
nlp_bc = en_ner_bc5cdr_md.load()
doc  = nlp_bc(sample)
# display its results
displacy_image = displacy.render(doc, jupyter =True, style = "ent")
```
